```python
import re
def cat(category):
    match = re.compile(r'Category:(\S+)?')
    return([m for m in match.findall(category)])
```

```python
df_new['category'] =df_new['wikicategories.value'].apply(cat)
```

```python
df_new.head()
```

```python
df_new['category'][2]
```

```python
def join_words(df):
    
    words_string= ','.join(df.category)

    return words_string

df_new['words_string'] = df_new.apply(join_words, axis=1)
print(df_new.head())
```

```python
# NEw dataframe pre-processing
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = df_new['abstract.value'].values.astype('U')
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
k = 5
model =KMeans(n_clusters=k,init='k-means++',max_iter=100,n_init=1)
model.fit(features)
```
